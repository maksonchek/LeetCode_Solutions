Последние опрераторы выполняют умножение на 2 (<<) целого числа, а (>>) - деление

```python
x = 10
y = 0
z = 0
```

```python
print(bin(x))
```

```python
y = x >> 1
print(y)
print(bin(y))
```

```python
z = x << 1
print(z)
print(bin(z))
```
